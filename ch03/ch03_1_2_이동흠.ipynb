**Chaper.3-1 단순선형회귀**

- 단순선형회귀 : 하나의 설명변수 X에 기초하여 양적 반응변수  Y를 예측하는 방법

- 이 기법의 특징은 X와 Y사이에 선형적 상관관계가 있다고 가정한다

- 이에 수학적으로 다음과 같이 나타낸다
  - $Y\approx\beta_{0}+\beta _{1}x$

- 여기에서 $\approx$은 근사적으로 모델링 된 것으로 볼 수가 있다.

- $\beta_{0}$ 과 $\beta_{1}$은 알려지지 않은 상수로 선형 모델의 절편과 기울기를 나타내며 또는 모델계수 또는 파라미터로 알려져 있다

- 이 두 가지는 알려지지 않은 상수로 선형 모델의 절편과 기울기를 나타내며 
이를 훈련 데이터를 이용하여 이에 수학적으로 선형적 상관관계는 다음과 같이 나타낸다.
- $\hat{y}$은 $X=x$ 일 때 Y의 예측값을 나타내며 이는 알려지지 않은 계수에 대한 추정값 또는 반응 변수의 예측값을 나타내는데 이용된다  


**Chaper.3.1.1 계수추정**

- 위에 나와 있는 $\beta_{0}$ 과 $\beta_{1}$ 알려져 있지 않기 때문에 예측하기 전에는 데이터를 이용해서 계수를 추정해야 한다

- 여기서는 선형모델이 이용 가능한 데이터에 잘 적합되도록 계수추정값인 $\beta_{0}$ 과 $\beta_{1}$ 을 얻는 것을 의미한다.

- 이를 알아보기 위해 가장 흔하게 사용하는 것이 최소제곱 기준을 최소로 하는 것이다.

- 만일 X의 i번째 값에 기초한 Y의 예측값을 $\hat{y}=\beta_{0}+\beta_{1}xi$ 라고 가정하면 다음과 같다 

- ei=${y}$-$\hat{y}i$는 i번째 잔차를 나타내며 이는 i번째 관측된 반응변수 값과 선형모델에 의해 예측된 i번째 반응변수의 차이가 된다  

  - 잔차제곱합(residual sum of squares)

**Chaper.3.1.2 계수 추정값의 정확도 평가**

- 실제 응용에서는 관측 자료를 이용하여서 최소제곱선을 계산할 수 있지만 모회귀선은 관측되지 않는데 아래의 보스턴 집값 데이터를 보게 되면 실제 모델을 사용 생성된 다른 데이터셋트는 다른 최소 제곱선을 가지고 있지만 모회귀선은 같은 모습을 보이고 있다.

- 물론 모회귀선과 최소제곱선 사이에 차이가 구별하기는 어렵지만 서로 다른 직선이 설명변수와 반응변수의 상관관계를 이야기 하는 것은 표본의 저옵를 사용 모집단의 특징을 추정하고자 하는 것이다

- 선형회귀와 확률변수의 평균값의 추정은 bias의 개념에서 볼 수가 있는데 평균적으로 $\hat{y}$이 평균적으로 y와 동일하다고 가정하는 것은 편향되지 않은것
으로 본다

- 물론 실제 데이터를 보게 되면 그리 되지 않을 수 있다 하지만 아주 많은 수의 관측치에서 얻은 추정값을 평균할수 있으면 값은 y와 정확하게 비슷한 값이 될 수가 있다 

- 즉 특정 데이터에서 $\beta_{0}$,$\beta _{1}$를 추정하면 그 추정값은 정확하게 일치하지 않으나 만일 아주 많은 수의 데이터세트에서 추정하면 추정값은 비슷하게 된다는 이야기가 된다

- 그럼 이러한 것에 있어서 추정값 a가 a의 추정값으로 얼마나 정확한지 알아 볼 수가 있는데 이를 우리는 표준오차로 계산한다 

  - 이러한 표준오차는 n이 증가 함에 따라 편차가 줄어드는지 아닌지를 말해주고 있는데 관측치가 많아질 수록 표준 오차는 작아지게 된다 그리고 이는 신뢰구간을 계산하는데도 이용이 되기도 한다
  - 표준오차는 계수들에[ 대한 가설 검정을 할 수 있는데 보통 사용되는 것은 귀무가설과 대립가설을 검정하며 그 모습은 다음과 같다
   - 가설 1 : X와 Y는 상관관계가 있다
   - 가설 2  : X와 Y는 상관관계가 없다
   

**Chaper.3.1.3 모델의 정확도 평가**

- 귀무 가설과 대립 가설을 통해 모델을 채택하면 모델이 얼마나 데이터에 적합한 정도를 수량화 할 필요가 있다
- 이에 보통 잔차표준오차(RSE)와  R2 통계량을 이용해서 평가한다
 - RSE : 데이터에 대한 모델의 적합성 결여를 나타내는 측도로 간주가 되며 모델을 사용하고 얻은 예측값이 실제 결과값에 가까우면 값은 작게 되고 실제 결과값에 가깝지 않으면 상당히 큰 값으로 나오기 때문에 데이터를 잘 적합하지 않는 다는 것을 알 수가 있다
  - R2 통계량 : X를 사용하여 설명될 수 있는 Y의 변동비율(Proportion of variabilty)를 측정하며 값이 1에 가까우면 반응변수의 변동 중 많은 부분이 회귀분석에 이용되었다는 것을 나타내며 0에 가까우면 변동의 대부분이 설명되지 않는다는 것을 알 수가 있다 
    - 물론 이 것은 다르게 이야기하면 correlation 즉 상관계수를 이용할 수가 있다는 이야기가 되며 상관계수의 제곱과 R2 통계량은 동일한 특징을 가지고 있다 

**Chaper.3.2.다중선형회귀**
- 우리가 어떠한 논문을 쓰거나 연구서를 읽다 보면 단일한 설명변수를 가지고 예측하는 일도 있지만 실제로는 여러가지 설명변수가 관련되게 되는데 이러한 1개 이상의 설명변수에 대한 분석을 어떻게 확장을 할지가 이번 파트의 핵심이다

**Chaper.3.2.1 회귀계수의 추정**
- 단순선형회귀분석과 같이 회귀계수는 $\beta_{0},\beta _{1},\beta _{2},\beta _{3}$...... 으로 알려지지 않은 값으로 되어 있으며 추정되어야 한다 이에 예측은 아래와 같은 식으로 되어 있다

- $\hat{y}$은 $\hat\beta_{0}$+$\hat\beta_{1}$+$\hat\beta_{2}$...... 

- $\beta_{0},\beta _{1},\beta _{2},\beta _{3}$......은 잔차제곱법을 최소화 하도록 선택한다 

- 파라미터를 표현하는 방식은 단순회귀에서 살펴보았던 것과 동일한 최소제곱법을 사용하여 추정, 잔차제곱합을 최소화 하도록 한다 


**Chaper.3.2.2 다중회귀분석에서 중요하게 생각해야 할 질문**

- 다중회귀분석에서는 변수들이 늘어나면서 생길 수 있는 질문이 있다
 - 1. 설명변수들이 적어도 하나는 반응변수를 예측하는데 유용하나
   - 반응변수와 설명변수 사이에 상관관계는 어떤가
     - 단순선형회귀에서는 베타의 값이 단순히 0 이면 결정할 수 있으나 하지만 설명변수가 여러게가 있는 경우 귀무가설과 대립가설을 설정 후 F- 통계량을 함으로 이뤄진다
     - 반응변수와 설명변수 사이에 상관관계가 없으면 F-통계량은 1에 가까운 수가 나올 것임
     - 반면 대립가설이 참이면 F-통계량의 기대값은 1보다 크게 될 것임
 - 2. Y를 설명하는데 모든 설명변수가 도움이 되는가 일부만이 되는가
     -   모든 설명변수가 반응변수와 상관성이 있을 수 있으나 실제로 연구를 보게 되면 일부만이 반응변수와 상관관계가 있게 된다 
     - 이에 어떤 모델이 가장 최고의 모델인가를 생각 할 수가 있는데
     이 목적을 위해서 효과적인 기법이 필요하게 된다
      1. 전진선택 : 절편만 포함하고 설명변수가 없는 모델을 가지고 설명 단순선형회귀를 적합하여 가장 RSS가 발생하는 변수를 영모델에 추가한 뒤 어떤 정지규칙을 만족할 때까지 계속한다
      2. 후진선택 : 모델의 모든 변수를 가지고 가장 큰 p값을 가지는 변수를 제외한 후 그 뒤 새로운 (p-1)의 변수를 적합하고 또 다른 p값을 가지는 변수를 제외한다
      3. 혼합선택 : 전진선택과 후진선택의 혼합
 - 3. 모델은 얼마나 잘 맞는가
   - 단순 회귀에서는 RSE와 R2의 값을 가지고 이야기 하나 다중선형회귀에서는 상관계수 (Cor Y, Y)의 제곱과 동일 한다 
   - 적합한 설명 변수는 상관계수가 최대가 된다는 것을 의미함 
- 4. 예측
  - 주어진 설명변수 값들에 대해 어떤 반응변수 값을 예측해야 하고 그 예측은 얼마나 정확한가?를 판단
  - 하지만 이러한 예측을 할 떄는 3가지 명확하지 않은 것이 존재함
   - 1.계수추정에 대한 추정값이 모회귀평면에 대한 추정값과 얼마나 유사한가?
   - 2.여기서는 항상 선형모델이 올바르다고 보기 때문에 예측이 정확하지 않을 수가 있다
   - 3.실제 값을 알아도 랜덤 오차 때무넹 반응변수 값을 정확히 예측할 수가 없다


 - 실제로 이러한 질문은 실제로 업무를 보거나(ex: 삼성전자의 다음날 시작가격이 어찌될 것인가?) 연구보고서를 작성하는데 가장 중요한 질문이 될 것이다

```
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

from scipy import stats
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```
# 평균을 저장하기 위한 변수선언
mean = 0

# 표준편차를 저장하기 위한 변수선언
std = 0

# 데이터의 값의 범의 [-1,1]를 일정하게 하기위한 함수
def normalize(X):
    X = (X - mean) / std  # 평균의 차를 표준편차로 나눔
    return X
```

```
from sklearn import datasets

# 보스턴 집값 데이터셋을 변수 boston_house_prices 변수에 저장
boston_house_prices = datasets.load_boston()

# 집값 데이터를 DataFrame타입으로 변환해서 data_frame변수에 저장
data_frame = pd.DataFrame(boston_house_prices.data)

# data_frame의 컬럼의 이름 설정 
data_frame.columns = boston_house_prices.feature_names
print(data_frame.tail())
```

```
# 집값이 저장된 target속성의 값을 price 컬럼을 추가해서 저장
data_frame['PRICE'] = boston_house_prices.target
print(data_frame.tail())
```

```
# 98
# 2개의 행과 4개의 열을 가진 subplots를 이용, axs는  4*2개의 ax를 가짐
fig, axs = plt.subplots(figsize = (16, 8), ncols=4, nrows=2)
lm_features = ['RM', 'ZN', 'INDUS', 'NOX', 'AGE', 'PTRATIO', 'LSTAT', 'RAD']
for i, feature in enumerate(lm_features):
    row = int(i/4)
    col = i % 4
    # 시본의 regplot을 이용해 산점도와 선형회귀 직선을 함께 표현
    sns.regplot(x = feature, y = 'PRICE', data = data_frame, ax = axs[row][col])
```
